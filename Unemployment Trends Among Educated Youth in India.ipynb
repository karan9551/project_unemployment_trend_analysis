```python
#histogram for numerical data
numeric_df = df[['LFPR', 'unemployment_rate', 'salary_expectation']]

numeric_df.hist(bins=20, figsize=(13, 6))
plt.suptitle('Histograms of LFPR, Unemployment Rate, and Salary Expectation', fontsize=12)
plt.savefig('../Project_Unemployment/output//unemployment_rate,LFPR and salary_exceptation.jpeg', dpi=100)
plt.show()

plt.title('Unemployment Rate Distribution')
plt.xlabel('Unemployment Rate (%)')
plt.ylabel('Number of Records')

plt.figure(figsize=(5, 6))
sns.heatmap(numeric_df.corr(), annot=False, cmap='coolwarm')
plt.title('Unemployment Rate Correlation Heatmap')  # Now placed correctly

plt.savefig('../Project_Unemployment/output//Unemployment Rate Distribution.jpeg', dpi=100)
plt.show()
```

```python
unemployment_rate_numaric_df=numeric_df
unemployment_rate_numaric_df.to_csv('../Project_Unemployment/output//unemployment_rate_LFPR_salary_exceptation.csv',index=False)
unemployment_rate_numaric_df.head(10)
```

```python
# Convert 'date' column to datetime format (if not already)
df['date'] = pd.to_datetime(df['date'])

# Extract 'year' from 'date'
df['year'] = df['date'].dt.year

#Group by 'year' and calculate average unemployment rate
# This gives us the overall unemployment trend each year
yearly_unemployment = (
    df.groupby('year')['unemployment_rate']
    .mean()
    .reset_index()
    .sort_values('year')
)

yearly_unemployment.head(10)
```

```python
# Group by year and calculate average unemployment rate
yearly_trend = df.groupby('year')['unemployment_rate'].mean().reset_index()

# Plot the trend
plt.figure(figsize=(10, 4))
sns.lineplot(data=yearly_trend, x='year', y='unemployment_rate', marker='o', color='teal')
plt.title('Average Unemployment Rate Among Educated Youth (2000–2024)', fontsize=13)
plt.xlabel('Year')
plt.ylabel('Unemployment Rate (%)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('../Project_Unemployment/output//Average Unemployment Rate Among Educated Youth (2000–2024).jpeg', dpi=100)
plt.show()

yearly_trend_unemloyment_among_edu_youth=yearly_trend
yearly_trend_unemloyment_among_edu_youth.to_csv('../Project_Unemployment/output//Average Unemployment Rate Among Educated Youth (2000–2024).csv',index=False)
yearly_trend_unemloyment_among_edu_youth.head(6)
```

```python
# State-wise average unemployment
state_avg = df.groupby('state')['unemployment_rate'].mean().sort_values(ascending=False).reset_index()

# Plot top 10 states with highest unemployment
plt.figure(figsize=(11, 5))
sns.barplot(data=state_avg.head(10), 
            x='unemployment_rate', 
            y='state',  
            hue='state',
            palette='mako',
            dodge=False,legend=False
           )
plt.title('Top 10 States with Highest Average Unemployment Rate (2000–2024)', fontsize=13)
plt.xlabel('Unemployment Rate (%)')
plt.ylabel('State')
plt.tight_layout()
# Save the most recent figure
plt.savefig('../Project_Unemployment/output//statewise_unemployment.jpeg', dpi=100)
plt.show()
state_avg_unemployment= state_avg
state_avg_unemployment.to_csv('../Project_Unemployment/output/statewise_unemployment.csv',index=False)
state_avg_unemployment
```

```python
edu_avg=df.groupby('education_level')['unemployment_rate'].mean().sort_values(ascending=False).reset_index()

plt.figure(figsize=(10, 4))
sns.barplot(data=edu_avg, x='education_level', y='unemployment_rate',hue='education_level', palette='coolwarm')
plt.title('Unemployment Rate by Education Level', fontsize=12)
plt.xlabel('Education Level')
plt.ylabel('Unemployment Rate (%)')
plt.xticks(rotation=40)
plt.tight_layout()
plt.savefig('../Project_Unemployment/output//Unemployment Rate by Education Level.jpeg', dpi=100)
plt.show()
edu_level_unemployment= edu_avg
edu_level_unemployment.to_csv('../Project_Unemployment/output/Unemployment Rate by Education Level.csv', index=False)
```
