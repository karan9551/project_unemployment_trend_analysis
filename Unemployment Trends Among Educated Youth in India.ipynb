```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Dataset
df = pd.read_csv(r"Data/unemployment.csv",encoding='latin1') #read csv file 
```

```python
df #show data in form of rows and coloumn by using pandas as pd
```

```python
#Show the first few rows
df.head(6)
```

```python
df.info()  #data types and nulls
df.describe() #statistcal summary
```

```python
print(df.columns) #list all columns
print("rows/column",df.shape) #show numbers of rows and column
```

```python
df.isnull().sum() #check null values and sum it 
#df.dropna(inplace=True) use this in case value in null 
print("Number of duplicates:",df.duplicated().sum()) #no duplicated values are present
# df.drop_duplicates(inplace = True) in case if there any duplicated values
```

```python
df[['LFPR', 'unemployment_rate', 'salary_expectation']].dtypes #the data in numeric form
```

```python
#Average Unemployment by Training Received
training_avg = df.groupby("training_received")["unemployment_rate"].mean()
plt.figure()
plt.pie(training_avg, labels=training_avg.index, autopct='%1.1f%%', startangle=100, colors=sns.color_palette("Set2"))
plt.title("Average Unemployment Rate by Training Received")
plt.axis('equal')
plt.tight_layout()
plt.savefig('../Project_Unemployment/output/Average Unemployment Rate by Training Received.jpeg', dpi=100)
plt.show()
```

```python
# Average Unemployment by Internet Access
internet_avg = df.groupby("internet_access")["unemployment_rate"].mean()
plt.figure()
plt.pie(internet_avg, labels=internet_avg.index, autopct='%1.1f%%', startangle=100, colors=sns.color_palette("Set1"))
plt.title("Average Unemployment Rate by Internet Access")
plt.axis('equal')
plt.tight_layout()
plt.savefig('../Project_Unemployment/output/Average Unemployment Rate by Internet Access.jpeg', dpi=100)
plt.show()
```

```python
#Average Unemployment by Dropout Reason
dropout_avg = df.groupby("dropout_reason")["unemployment_rate"].mean().sort_values(ascending=False)
plt.figure()
plt.pie(dropout_avg, labels=dropout_avg.index, autopct='%1.1f%%', startangle=100, colors=sns.color_palette("Set3"))
plt.title("Average Unemployment Rate by Dropout Reason")
plt.axis('equal')
plt.tight_layout()
plt.savefig('../Project_Unemployment/output/Average Unemployment Rate by Dropout Reason.jpeg', dpi=100)
plt.show()
```

```python
#histogram for numerical data
numeric_df = df[['LFPR', 'unemployment_rate', 'salary_expectation']] #selecting columns from df Histogram of 3 column

# Histograms of numeric columns 
numeric_df.hist(bins=20, figsize=(13, 6)) 
plt.suptitle('Histograms of LFPR, Unemployment Rate, and Salary Expectation', fontsize=12) # sub tittle 
plt.savefig('../Project_Unemployment/output//unemployment_rate,LFPR and salary_exceptation.jpeg', dpi=100)
plt.show()

#Optional: Individual Histogram for Unemployment Rate only
plt.figure(figsize=(6, 4)) # Creates a new blank figure
plt.hist(
    df['unemployment_rate'], 
    bins=20, 
    color='skyblue',
    edgecolor='black'
    ) # Creates a new blank figure with a size of 6 inches wide and 4 inches tall
plt.title('Unemployment Rate Distribution')
plt.xlabel('Unemployment Rate (%)')
plt.ylabel('Number of Records')
plt.savefig('../Project_Unemployment/output/unemployment_rate_distribution.jpeg', dpi=100) #save img file in output folder
plt.show()

# show Correlation Heatmap
plt.figure(figsize=(5, 6))
sns.heatmap(numeric_df.corr(),
            annot=False, 
            cmap='coolwarm'
           ) #Computes the correlation matrix (values from -1 to +1) between numerical columns like LFPR, unemployment_rate, salary_expectation
plt.title('Unemployment Rate Correlation Heatmap')  # tittle 

plt.savefig('../Project_Unemployment/output/unemployment_rate_correlation_heatmap.jpeg', dpi=100)
plt.show()
```

```python
# Convert 'date' column to datetime format (if not already)
df['date'] = pd.to_datetime(df['date'])

# Extract 'year' from 'date'
df['year'] = df['date'].dt.year
```

```python
#Group by 'year' and calculate average unemployment rate
# This gives us the overall unemployment trend each year
yearly_unemployment = (
    df.groupby('year')['unemployment_rate']
    .mean()
    .reset_index()
    .sort_values('year')
)
yearly_unemployment.head(6)
```

```python
# Group by yearly trend and calculate average unemployment rate
yearly_trend = df.groupby('year')['unemployment_rate'].mean().reset_index()

# Plot the trend
plt.figure(figsize=(10, 4))
sns.lineplot(data=yearly_trend, x='year', y='unemployment_rate', marker='o', color='teal')
plt.title('Average Unemployment Rate Among Educated Youth (2000–2024)', fontsize=13)
plt.xlabel('Year')
plt.ylabel('Unemployment Rate (%)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('../Project_Unemployment/output/Average Unemployment Rate Among Educated Youth (2000–2024).jpeg', dpi=100)
plt.show()


yearly_trend_unemloyment_among_edu_youth=yearly_trend.head()
# to save csv file output
yearly_trend_unemloyment_among_edu_youth.to_csv('../Project_Unemployment/csv_output/Average Unemployment Rate Among Educated Youth (2000–2024).csv',index=False)
#yearly_trend_unemloyment_among_edu_youth.head(5)
#yearly_trend_unemloyment_among_edu_youth.tail(5)
```

# Top 10 States with Highest Average Unemployment Rate (2000–2024)

```python
# State-wise average unemployment
state_avg = df.groupby('state')['unemployment_rate'].mean().sort_values(ascending=False).reset_index()

# Plot top 10 states with highest unemployment
plt.figure(figsize=(11, 5))
sns.barplot(data=state_avg.head(10), 
            x='unemployment_rate', 
            y='state',  
            hue='state',
            palette='mako',
            dodge=False,
            legend=False
           )
plt.title('Top 10 States with Highest Average Unemployment Rate (2000–2024)', fontsize=13)
plt.xlabel('Unemployment Rate (%)')
plt.ylabel('State')
plt.tight_layout()
# Save the most recent figure
plt.savefig('../Project_Unemployment/output//statewise_average Unemployment Rate (2000–2024).jpeg', dpi=100)
plt.show()


state_avg_unemployment= state_avg
# to save csv file
state_avg_unemployment.to_csv('../Project_Unemployment/csv_output/statewise_average Unemployment Rate (2000–2024).csv',index=False)
state_avg_unemployment.head(5)
```

# Unemployment Rate by Education Level

```python
#Unemployment Rate by Education Level
edu_avg=df.groupby('education_level')['unemployment_rate'].mean().sort_values(ascending=False).reset_index()

plt.figure(figsize=(8, 4))
sns.barplot(data=edu_avg, x='education_level', y='unemployment_rate',hue='education_level', palette='coolwarm')
plt.title('Unemployment Rate by Education Level', fontsize=12)
plt.xlabel('Education Level')
plt.ylabel('Unemployment Rate (%)')
plt.xticks(rotation=40)
plt.tight_layout()
plt.savefig('../Project_Unemployment/output/Unemployment Rate by Education Level.jpeg', dpi=100)
plt.show()

edu_level_unemployment= edu_avg
# to save csv file
edu_level_unemployment.to_csv('../Project_Unemployment/csv_output/Unemployment Rate by Education Level.csv',index=False)
edu_level_unemployment
```

# Average Gender level unemployment rate

```python
# genderlevel unemployment
gender_avg=df.groupby('gender')['unemployment_rate'].mean().sort_values(ascending=False).head()

plt.figure(figsize=(4, 4))
plt.pie(
        gender_avg, 
        labels=gender_avg.index, 
        autopct='%1.1f%%', 
        startangle=120, 
        colors=sns.color_palette('Paired')
       )
plt.title('Share of Average Unemployment Rate by Gender', fontsize=12)
plt.axis('equal')
plt.tight_layout()
plt.savefig('../Project_Unemployment/output/pie chart-Share of Average Unemployment Rate by Gender.jpeg', dpi=100)
plt.show()
gender_level_unemployment = gender_avg.reset_index()
#to save csv output
gender_level_unemployment.to_csv('../Project_Unemployment/csv_output/pie chart-Share of Average Unemployment Rate by Gender.csv', index=False)
gender_level_unemployment
```

```python
gender_avg = df.groupby('gender')['unemployment_rate'].mean().reset_index()
gender_avg.columns = ['Gender', 'Average Unemployment Rate']

# Plot
plt.figure(figsize=(4, 3))
sns.barplot(
    data=gender_avg,
    x='Gender',
    y='Average Unemployment Rate',
    hue='Gender',
    palette='Paired',
    legend=False
)
plt.ylabel('Average Unemployment Rate (%)')
plt.title('Average Unemployment Rate by Gender')
plt.tight_layout()
plt.savefig('../Project_Unemployment/output/Bar Chart - Average Unemployment Rate by Gender.jpeg', dpi=100)
plt.show()
```
